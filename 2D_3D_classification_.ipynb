```python
def check_black_rate(img):
    try:
        label = 0
        count = 0
        w,h,_ = img.shape
        for j in range(w):
            for i in range(h):
                if (img[j][i] < [28,28,28]).all():
          
                    count += 1
      
        entire = img.size/3
        black_rate = count / entire * 100
        print(math.trunc(black_rate),"%")
    
        if black_rate <= 68:
            label = 3
            #print(f"{label}d폴더로 이동")
        else:
            label = 2
            #print(f"{label}d 폴더로 이동")
            
        return label
    finally:
        count = 0
```

```python
def noise_and_edge(img_path):
    bdn = cv2.imread(img_path)
    img = cv2.fastNlMeansDenoisingColored(bdn, None, 10,10,7,21)
    
    gx_kernel = np.array([[-3,0, 3], [-10,0,10], [-3,0,3]])
    gy_kernel = np.array([[-3,-10,-3],[0,0,0], [3,10,3]])

    edge_gx = cv2.filter2D(img, -1, gx_kernel)
    edge_gy = cv2.filter2D(img, -1, gy_kernel)

    edge = edge_gx + edge_gy
    
    pix = np.array(edge)
    return pix
    
```

```python
import cv2
import numpy as np
import os, glob
import math
import shutil
import re

label = 0
path = '/Users/kimminyoung/Desktop/2d_3d_분류_폴더구조수정'

#D:/data/data_cls/2D
#D:/data/data_cls/3D

os.makedirs(f'{path}/data_cls/2D', exist_ok = True)
os.makedirs(f'{path}/data_cls/3D', exist_ok = True)
data_path = f'{path}/data_test/final_crop'

for ani_folder in os.listdir(data_path):
    
    #.DS_Store 무시
    if ani_folder.startswith("."):
        
        continue 
    else:
        print(ani_folder) #애니메이션명
    
    #이동시킬 폴더
    move_folder = data_path + f'/{ani_folder}'
    
    for id_folder in os.listdir(data_path + f'/{ani_folder}'):
        

        #.DS_Store 무시
        if id_folder.startswith("."):
            continue
        else:
            print(id_folder) #캐릭터id명
            
        
        for crop_folder in os.listdir(data_path + f'/{ani_folder}/{id_folder}'):
            #print(crop_folder)

            #.DS_Store 무시
            if crop_folder.startswith("."):
                continue
                
            
            for file in os.listdir(data_path + f'/{ani_folder}/{id_folder}/{crop_folder}'):
                #print(file)

                #wav파일이면 무시    
                if file.endswith(".wav"):
                    continue
                
                #.DS_Store 무시
                if file.startswith("."):
                    continue
                
                edge = noise_and_edge(data_path + f'/{ani_folder}/{id_folder}/{crop_folder}/{file}')

                label = check_black_rate(edge)

                if label == 2:
                    desti_2d = path + f'/data_cls/2D/{ani_folder}/'
                    shutil.move(move_folder, desti_2d)
                    
                else:
                    desti_3d = path + f'/data_cls/3D/{ani_folder}/'
                    shutil.move(move_folder, desti_3d)
                    
                print(f"{label}d폴더로 이동")
                print("\n")

                break
            
            else:
                continue
            
            break
            
        break
        
```
